```python
plt.figure(figsize=(4,3))
my_labels=['Survived','Not Survived']
my_explode=[0.05,0]
plt.pie(x=df['Survived'].value_counts(),labels=my_labels,autopct='%0.2f%%',explode=my_explode)
plt.title('Survived Status')
centre_circle = plt.Circle((0, 0), 0.40, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
m=sns.countplot(x='Sex',data=df)
m.bar_label(m.containers[0])
plt.subplot(1,2,2)
m=sns.countplot(x='Sex',hue='Survived',data=df)
m.bar_label(m.containers[0])
m.bar_label(m.containers[1])
```

```python
plt.figure(figsize=(6, 3))
sns.histplot(data=df, x='Age', hue='Survived',multiple='stack', kde=True, palette='Set1')
plt.title('Distribution of Survival by Age')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.show()
```

```python
# encoding
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['Sex']=label.fit_transform(df['Sex'])
df['Embarked']=label.fit_transform(df['Embarked'])
df
```

```python
x=df.drop(['Survived'],axis=1)
x
```

```python
y=df['Survived'].values

y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_train
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train, y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#accuracy_score=TP/(TP+TN+FP+FN)
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
result=confusion_matrix(y_test,y_pred)
score=accuracy_score(y_test,y_pred)
print(score)
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.metrics._plot.confusion_matrix import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=['1','0']
cmd=ConfusionMatrixDisplay(result,display_labels=cm)
cmd.plot()
```
